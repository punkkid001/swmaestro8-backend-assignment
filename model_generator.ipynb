```python
import sys
import json
import gensim
import konlpy
import codecs
import multiprocessing

from konlpy.tag import Kkma, Twitter
from gensim.models import Word2Vec
```

```python
tagger = Kkma()
with open("refined_category_dataset.dat") as fin:
    sentences = []
    for line in fin.readlines():
        info = json.loads(line.strip())
        
        word = tagger.morphs(info['name'].replace('/', ' ').replace('(', '') .replace(')', '').replace('[', '').replace(']', '').replace('#', ''))
        word2 = tagger.nouns(info['name'].replace('/', ' ').replace('(', '') .replace(')', '').replace('[', '').replace(']', '').replace('#', ''))
        
        word.append(info['cate'])
        word2.append(info['cate'])
        
        sentences.append(word)
        sentences.append(word2)
```

```python
print(sentences[:100])

config = {
    'min_count': 5,
    'size': 100,
    'sg': 1,
    'batch_words': 10000,
    'iter': 10,
    'workers': multiprocessing.cpu_count(),
}

model = gensim.models.Word2Vec(**config)
model.build_vocab(sentences)
model.train(sentences, total_examples=model.corpus_count, epochs=model.iter)

#model = KeyedVectors.load('ko.bin')
```

```python
model.save('word_models.model')
```
